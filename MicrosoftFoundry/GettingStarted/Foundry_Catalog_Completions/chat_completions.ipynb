### Importing Environment Variables

```python
import os
from pathlib import Path
from dotenv import load_dotenv
from anthropic import AnthropicFoundry

# Look for the .env file next to this notebook, even if the kernel starts in a different folder
candidates = [
    Path.cwd() / ".env",
    Path.cwd() / "MicrosoftFoundry" / "GettingStarted" / "Foundry_Catalog_Completions" / ".env",
    Path.cwd().parent / "MicrosoftFoundry" / "GettingStarted" / "Foundry_Catalog_Completions" / ".env",
]

env_path = next((p for p in candidates if p.exists()), None)

if env_path is None:
    raise FileNotFoundError("Could not find the .env file. Create one next to this notebook with MODEL_ENDPOINT, MODEL_DEPLOYMENT_NAME, and API_KEY.")

load_dotenv(dotenv_path=env_path)

model_endpoint = os.getenv("MODEL_ENDPOINT")
model_deployment_name = os.getenv("MODEL_DEPLOYMENT_NAME")
api_key = os.getenv("API_KEY")

if not model_endpoint or not model_deployment_name or not api_key:
    raise ValueError("One or more required values are missing. Update the .env file with your Foundry endpoint, deployment name, and API key, then rerun this cell.")
```

### Setting Up Anthropic Client

```python
client = AnthropicFoundry(
    api_key=api_key,
    base_url=model_endpoint,
)
```

### Sending a Request to the Model

```python
message = client.messages.create(
    model = model_deployment_name,
    system = "You are a helpful assistant that translates English to Spanish.",
    messages = [
        {
            "role": "user",
            "content": "Translate the following English text to Spanish: 'Hi, my name is Kuljot. How are you today?'"
        }
    ],
    max_tokens=1024
)

json_message = message.to_dict()
print(json_message['content'][0]['text'])
```
